# Power

# Continued Fractions

```python
def list_to_frac(l):
    expr = Integer(0)
    for i in reversed(l[1:]):
        expr += i
        expr = 1/expr
    return l[0] + expr
list_to_frac([x, y, z])
```

```python
list_to_frac([1, 2, 3, 4])
```
