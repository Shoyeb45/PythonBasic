```python
from sympy import *
```

```python
x, y, z = symbols('x y z')
init_printing()
```

# Derivatives

```python
diff(cos(x), x)
```

```python
diff(cos(x), y)
```

```python
diff(exp(x), x)
```

```python
diff(exp(x**3), x)
```

## Multiple derivatives

```python
expr = x**4
```

```python
diff(expr, x, x) # double differentiation
```

```python
diff(expr, x, 4)
```

```python
expr = exp(x*y*z)
```

```python
diff(expr, x, y, y, z, z, z, z)
```

```python
diff(expr, x, y, y, z, 4)
```

```python
expr_derivative = Derivative(expr, x, y, 2, z, 4)
expr_derivative
```

```python
expr_derivative.doit() # Evaluate
```

```python
m, n, a, b = symbols('m n a b')
expr = (a*x + b)**m
expr.diff((x, n))
```

# Integrals

```python
integrate(cos(x), x)
```

# Numeric Integration

```python
integral = Integral(sqrt(2)*x, (x, 0, 1))
integral
integral.evalf()
```

```python
Integral(1 / sqrt(x), (x, 0, 1)).evalf()
```

# Series Expansion

```python
expr = exp(sin(x))
expr.series(x, 0, 4)
```

- The term at the end represents the Landau order term at (not to be confused with big O notation used in computer science, which generally represents the Landau order term at where). It means that all x terms with power greater than or equal to are omitted. Order terms can be created and manipulated outside of series. They automatically absorb higher order terms.

```python
expr.series(x, 0, 10).removeO()
```

```python
expr = exp(x)
expr.series(x, 0, 4)
```

```python
exp(x - 6).series(x, x0=6)
```
