```python
from sympy import *
from sympy.abc import *
```

# Substitution

```python
expr = cos(x) + 1
```

```python
expr.subs(x, psi)
```

```python
expr = x ** y
expr
```

```python
expr = expr.subs(y, x ** y)
expr
```

```python
expr = expr.subs(y, x ** y)
expr
```

```python
expr = cos(2 * x) + sin(2 * x)
expr
```

```python
expand_trig(expr)
```

```python
# we don't want cos(2x) to be expanded
expr.subs(sin(2 * x), 2 * sin(x) * cos(x))
```

# Converting strings to sympy symbols

```python
str_exprs = 'x ** 2 + 3 * x - sin(pi*x)'
exprs = sympify(str_exprs)
exprs
```

# `lambdify` function

```python
exprs = 1 / (1 + exp(x))
exprs
```

```python
sigma = lambdify(x, exprs)
```

```python
import numpy as np
a = np.arange(10)
sigma(a)
```
